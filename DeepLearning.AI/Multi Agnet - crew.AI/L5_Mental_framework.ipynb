

there is a very high correlation between being a good manager and being able to create great multi-agent systems, because managers are conditioned to think about what is the goal and what is the process.



agents do a better job if they have a few things: 
- they are doing role-playing. 
- They have focus. 
- They have the appropriate tools. 
- They're allowed to cooperate. 
- They have some guardrails. 
- They also can leverage memory.
