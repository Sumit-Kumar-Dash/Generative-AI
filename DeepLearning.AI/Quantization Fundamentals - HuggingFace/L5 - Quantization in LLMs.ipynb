















    Fine tuning a quantized models are of two types:
        1.fine tune a model while being quantized to get the best quantized model possible
        2.adapt their model for specific use case, fine tuning an LLM on their own datset


    1. The first scenario is doable through Quantization Aware Training -> we train the model to be more accurate once we quantize it


    2.For second use case we would use PEFT -> Parameter Efficient Fine Tuning methods . It trys to frastically reduce the no. of trainable parameters of the model while trying to keep the same performance as full tuning.
    PEFT + QLorA => both quantization & PEFT methods

#### LoRa (Low Rank Adapters) =>


    When doing LoRa you simply have to attach extra trainable parameters (blue color) to a frozen parameters (orange color).
    Since r is extremly small ususally compared to input hidden state dimensions, so the final optimizer state end up being extremly small which making the training protocol much more accessible



    QLoRa leverage this concept by quantizing base weights (blue color) in 4 bit precision and making sure data type of activation of quantized weight matches the data type of the LoRa weights.That way we can perform sum easily of pre trained (blue) and LoRa weights (orange)
