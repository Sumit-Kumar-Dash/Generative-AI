# Beyond Linear Quantization

    Outlier feature means hiddens state with large magnitude
    features predicted by the models means magnitude of hidden states satrted to get large thus making quantization quite obsolete by failing on LLM models

- LLM.int8() propose decompose the underlying matrix multiplication of thr linear layer in two stages. 
- Decompose the i/p into all the hidden states that are greater than certain threshold and the non outlier part matrix multiplication in int 8.
- So in quantize do the matmul in 8 bit and then dequantize using scales to get the final result in i/p datatype.
- Second part, do classical matmul with original dtype of hidden state thats in half precision. 
- Then combine both results . This way you can retain the full performance of the model without any performance degradation .
