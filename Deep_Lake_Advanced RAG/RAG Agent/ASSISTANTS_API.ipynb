```python
# If run is 'completed', get messages and print
while True:
  # Retrieve the run status
  run_status = client.beta.threads.runs.retrieve(thread_id=thread.id,run_id=run.id)
  print(run_status.model_dump_json(indent=4))
  time.sleep(10)
  if run_status.status == 'completed':
    messages = client.beta.threads.messages.list(thread_id=thread.id)
    print(messages)
    break
  else:
    ### sleep again
    time.sleep(2)
```

```python
from transformers import ViTImageProcessor, ViTForImageClassification
from PIL import Image
import requests

url = 'http://images.cocodataset.org/val2017/000000039769.jpg'
image = Image.open(requests.get(url, stream=True).raw)

processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')

inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
# model predicts one of the 1000 ImageNet classes
predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```

```python
import requests

# Function to fetch text from the API
def fetch_text_from_api(url):
    api_url = "https://lexper.p.rapidapi.com/v1.1/extract"
    querystring = {"url": url, "js_timeout": "30", "media": "true"}
    headers = {
        "X-RapidAPI-Key": "xxx",
        "X-RapidAPI-Host": "lexper.p.rapidapi.com"
    }
    response = requests.get(api_url, headers=headers, params=querystring)
    data = response.json()
    # Adjust this line according to the structure of your API response
    return data.get('article', {}).get('text', '')

# Function to summarize the text using Hugging Face API
def query_huggingface(text):
    if not text.strip():  # Check if text is empty or only contains whitespace
        return {"error": "No text to summarize"}

    API_URL = "https://api-inference.huggingface.co/models/facebook/bart-large-cnn"
    headers = {"Authorization": f"Bearer {API_TOKEN}"}
    payload = {"inputs": text, "parameters": {"do_sample": False}}
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

while True:
    url_to_summarize = input("Enter the URL to summarize or 'exit' to quit: ")

    if url_to_summarize.lower() == 'exit':
        break

    text_to_summarize = fetch_text_from_api(url_to_summarize)
    summary_response = query_huggingface(text_to_summarize)
    print(summary_response)
```
